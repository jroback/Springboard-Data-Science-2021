## The Elements of Style

What makes up a style?   
It's less about what it it has in common with other works, and more about what makes it unique.   
In other words, the distinguishable traits unique from similar subject matter.

This notebook explores the characteristics of the Berkley-Monet dataset—1150 impressionist paintings by Monet.   
The goal is to explore the images to find what the data say makes a Monet, a Monet.   
We'll do this by comparing the components of Monet's work with the components of a modern artist known for his landscape work—Bob Ross.  
When we take the visual component out, can we still discern it's a Monet?   
Let's find out.

### Imports and Loading the Data

```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import cv2
from PIL import Image
import sys
import skimage.io
import skimage.viewer
from tensorflow import keras
from tensorflow.keras.models import Sequential, Model
from  matplotlib import pyplot as plt
import matplotlib.image as mpimg
from matplotlib.pyplot import figure
```

```python
IMG_WIDTH=256
IMG_HEIGHT=256
img_folder="data/"

def create_dataset_PIL(img_folder):
    img_data_array=[]
    for file in os.listdir(os.path.join(img_folder)):
        image_path = os.path.join(img_folder, file)
        image = np.array(Image.open(image_path))
        image = np.resize(image,(IMG_HEIGHT, IMG_WIDTH, 3))
        image = image.astype('float32')
        image /= 255  
        img_data_array.append(image)
    return img_data_array

PIL_img_data = create_dataset_PIL(img_folder)
```

```python
test_image = PIL_img_data[2]
plt.imshow(test_image)
plt.show()
```

```python
test_image2 = PIL_img_data[20]
plt.imshow(test_image2)
plt.show()
```

Monet shows such uniquness in the brush strokes, the colors and the overall mood.   
But can this be seen in the data behind the images?

### Not Monet

```python
#Resize images
path = "not_monet/"
dirs = os.listdir( path )

def resize():
    for item in dirs:
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((IMG_HEIGHT, IMG_WIDTH), Image.ANTIALIAS)
            imResize.save(f + ' resized.jpg', 'JPEG', quality=90)

resize()
```
